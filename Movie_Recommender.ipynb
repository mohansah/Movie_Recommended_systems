```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Reading ratings file:
r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings=pd.read_csv("u.data",names=r_cols,sep='\t')

print("Rating Informations")
print(ratings.shape)
ratings.head()
```

```python
user_df=pd.DataFrame(ratings['user_id'])
print(user_df.shape)

user_df.drop_duplicates(inplace=True)
print(user_df.shape)
user_df.head(10)
```

```python
#Reading items file:
i_cols = ['movie_id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure','Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('u.item', sep='|',encoding='latin-1', names=i_cols)

print("Rating Informations")
print(items.shape)
items.head()
```

```python
movie_df=pd.DataFrame(items[['movie_id','movie title']])
print(movie_df.shape)

movie_df.drop_duplicates(inplace=True)
print(movie_df.shape)
movie_df.head(10)
```

```python
df = pd.merge(ratings, items, on='movie_id')
print(df.shape)
df.head(5)
```

```python
len(df['movie_id'].unique())
```

```python
len(df['movie title'].unique())
```

# Simple popularity-based recommender based on Number of Views

```python
movie_df1=pd.DataFrame(df.groupby(['movie_id','movie title'])['user_id'].count())
movie_df1.head()
```

```python
movie_df1['Number_Of_Views']=movie_df1['user_id']
movie_df1=movie_df1.drop('user_id',axis=1)
movie_df1.head()
```

# Showing Popular Movies on the basis of Number of Views

```python
popular_movie1=movie_df1.sort_values(by='Number_Of_Views',ascending=False)
popular_movie1['Rank']=range(1,len(popular_movie1)+1)
popular_movie1.head(10)
```

```python
pd.merge(movie_df,popular_movie1,on="movie_id").sort_values(by='Number_Of_Views',ascending=False).head(15)
```

# Simple popularity-based recommender based on Ratings

```python
movie_df2=pd.DataFrame(df.groupby(['movie_id','movie title'])['rating'].mean())
movie_df2.head()
```

```python
movie_df2['Number_Of_Rating']=df.groupby(['movie_id','movie title'])['rating'].count()
movie_df2.head()
```

```python
popular_movies2=movie_df2.sort_values(by='rating',ascending=False)
popular_movies2.head(10)
```

# Showing Popular Movies on the basis of Ratings

There is possibility that there is only one rating for a movie and its value is 5 in that case suggesting that movie is not correct so we have to decide a threshold value for number of rating

```python
# suppose threshold = 100

popular_movies2_final=popular_movies2[popular_movies2['Number_Of_Rating']>=100]
popular_movies2_final['Rank']=range(1,len(popular_movies2_final)+1)
popular_movies2_final.head(10)
```

```python
pd.merge(movie_df,popular_movies2_final,on="movie_id").sort_values(by='rating',ascending=False).head(10)
```

# Build a Movie recommender with personalization

# Item Based

```python
movie_matrix2 = df.pivot_table(index='user_id', columns='movie_id', values='rating')
movie_matrix2.head(10)
```

```python
# we are finding movies similar to a particular movie

movie=20

movie_ratings=movie_matrix2[movie]
movie_ratings.head(10)
```

```python
similar_to_the_movie=movie_matrix2.corrwith(movie_ratings)
print(similar_to_the_movie.shape)
similar_to_the_movie.head(10)
```

```python
movie_record=pd.DataFrame(similar_to_the_movie,columns=['Correlation'])
movie_record.head(15)
```

```python
movie_record=movie_record.join(popular_movies2['Number_Of_Rating'])
movie_record.head(15)
```

```python
movie_record.dropna(inplace=True)
movie_record.head(10)
```

```python
suggested_movies=movie_record.sort_values(by='Correlation',ascending=False)
suggested_movies.head(15)
```

```python
# suppose threshold=100

suggested_movies_final=suggested_movies[suggested_movies['Number_Of_Rating']>100]
suggested_movies_final['Rank']=range(1,len(suggested_movies_final)+1)
suggested_movies_final.head(20)
```

```python
pd.merge(movie_df,suggested_movies_final,on="movie_id").sort_values(by='Correlation',ascending=False).head(10)
```

# User Based

```python
movie_matrix3 = df.pivot_table(index='movie_id', columns='user_id', values='rating')
movie_matrix3.head(10)
```

```python
# we are finding user similar to a particular user

user=20

user_ratings=movie_matrix3[user]
user_ratings.head(10)
```

```python
similar_to_the_user=movie_matrix3.corrwith(user_ratings)
print(similar_to_the_user.shape)
similar_to_the_user.head(10)
```

```python
user_record=pd.DataFrame(similar_to_the_user,columns=['Correlation'])
user_record.head(10)
```

```python
user_record.dropna(inplace=True)
suggested_users=user_record.sort_values(by='Correlation',ascending=False)
suggested_users.head(20)
```

```python
suggested_users=pd.merge(user_df,suggested_users,on="user_id").sort_values(by='Correlation',ascending=False)
suggested_users.head(20)
```

```python
# List of users which are similar to the that particular user

suggested_users['Rank']=range(1,len(suggested_users)+1)
suggested_users.head(20)
```
